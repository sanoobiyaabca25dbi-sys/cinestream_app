```python
df = df.drop_duplicates().copy()
```

```python
text_cols = [
    "Type",
    "Genre",
    "Language",
    "Country",
    "Director",
    "AgeRating",
    "TrendingStatus"
]

for col in text_cols:
    df[col] = df[col].astype(str).str.strip()
```

```python
df["Type"] = df["Type"].str.title()
```

```python
df["AddedDate"] = pd.to_datetime(
    df["AddedDate"],
    errors="coerce"
)
```

```python
df["Profit_Cr"] = (
    df["RevenueCr"]
    - df["ProductionCostCr"]
)
```

```python
df["ROI_Pct"] = (
    df["Profit_Cr"]
    / df["ProductionCostCr"]
) * 100
```

```python
df["Performance_Band"] = np.where(
    df["Profit_Cr"] > 20,
    "Hit",
    np.where(
        df["Profit_Cr"] >= 0,
        "Break-even",
        "Flop"
    )
)
```

```python
df.head()
```

```python
df.loc[(df['IMDbScore'] < 1.0) | (df['IMDbScore'] > 10.0), 'IMDbScore'] = np.nan
```

```python
df.loc[(df['CriticRating'] < 1.0) | (df['CriticRating'] > 10.0), 'CriticRating'] = np.nan
```

```python
df.loc[df['SubscribersGainedThousands'] < 0, 'SubscribersGainedThousands'] = 0
```

```python
df.loc[df['RuntimeMinutes'] > 1000, 'RuntimeMinutes'] = np.nan
```

```python
df.to_csv("../outputs/cleaned_cinestream.csv",index=False)
```
